### 6.Matplotlib Basics:
○ Plot a simple line chart.
○ Create a scatter plot

```python
import matplotlib.pyplot as plt
import random

#generating random data and plotting a line graph
x,y = [0] , [0]
for i in range(100):
    x.append(random.randrange(x[-1],x[-1]+10))
    y.append(random.randrange(y[-1],y[-1]+10))


plt.plot(x, y, color='g', label="Random data")
plt.xlabel("random x values")
plt.ylabel("random y values")
plt.title("basic line graph")
plt.legend()
plt.show()

#generating random data and plotting a scatter plot
x , y = [] , []
x,y = [0] , [0]
for i in range(100):
    x.append(random.randrange(1,100))
    y.append(random.randrange(30,100))
plt.scatter(x,y)
plt.xlabel("random x values")
plt.ylabel("random y values")
plt.title("scatter plot")
plt.show()
```

### 7. Data Visualization with Seaborn:
○ Create a box plot to visualize data distributions.○ Generate a histogram and a density plot

```python
#using dataframe sentiment.csv to make the plots
df = pd.read_csv("sentiment.csv");

#fixing the data
df["Text"] = df["Text"].str.strip().str.lower()
df["Sentiment"] = df["Sentiment"].str.strip().str.lower()

#plotting the text length, per sentiment as a box plot
sentiments = df['Sentiment'].value_counts()
df['Text_Length'] = df['Text'].str.len()
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sentiment', y='Text_Length', order=sentiments.index[0:20],palette = "viridis")
plt.title('Text Length for Each Sentiment Category')
plt.xlabel('Sentiment')
plt.ylabel('Text Length')
plt.xticks(rotation=45)
plt.show()

#plotting text lengths as a histrogram
df['Text_Length'] = df['Text'].str.len()
plt.figure(figsize=(12, 6))
sns.histplot(data = df ,x = "Text_Length" ,palette = "viridis")
plt.title('Length of various texts')
plt.ylabel('No of texts')
plt.show()

# plotting text length as density plot
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x="Text_Length", fill=True, color="blue")
plt.title("density plot of length of texts")
plt.xlabel("length")
plt.ylabel("density")
plt.show()
```

### 8. Time Series Data: 
○ Plot a time series using Pandas and Matplotlib

```python
#random data generated using chatgpt
data = {"Date": pd.date_range(start="2020-01-01", periods=30, freq="D"),
    "Bday_count": [random.randint(1,200) + i  for i in range(30) ] } # Random pattern

#converting to dataframe
df = pd.DataFrame(data)
display(df)

#converting to date time and setting index
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

# Plotting graph
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["Bday_count"], marker="o", linestyle="-", color="g")
plt.title("Birthdays on a given date")
plt.xlabel("Date")
plt.ylabel("Bday_count")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

### 9. Correlation Analysis:
○ Generate a correlation heatmap using Seaborn.

```python
# random data from chatgpt
data = {
    "Age": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "Salary": [3000, 4000, 5000, 6000, 8000, 11000, 15000, 20000, 25000, 30000],
    "Experience": [1, 3, 5, 8, 10, 12, 15, 20, 25, 30],
    "Work Hours": [40, 42, 44, 45, 43, 41, 40, 38, 37, 35]
}

#creating the dataframe and the matrix
df = pd.DataFrame(data)
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("corr heatmap")
plt.show()
```

### 10. Data Aggregation:
○ Group data using Pandas and visualize the results

```python
#back to using sentiments.csv 
df = pd.read_csv("sentiment.csv")
#grouping data by sentiment and text length of sentiments
df['Text_Length'] = df['Text'].str.len()
text_length_by_sentiment = df.groupby('Sentiment')['Text_Length'].mean().head(10)
print("\nAverage Text Length by Sentiment:")
display(text_length_by_sentiment)

plt.figure(figsize=(12, 6))
sns.barplot(y=text_length_by_sentiment.index, x=text_length_by_sentiment.values,palette = "viridis")
plt.title('Average Text Length by Sentiment Category')
plt.xlabel('Sentiment')
plt.ylabel('Average Text Length')
plt.show()
```

### 11. Data Cleaning:
○ Handle missing values and perform data imputation.

```python
#the following is a question from DAV. Using the question and dataset as it is relavent to task 11
'''2. For the CEREALS dataset, perform data preprocessing and answer the following questions. 
a. Create a table with the 5 number summary of all the numeric attributes.
b. For each of the numeric attributes (proteins upto vitamins) , identify and replace all missing data(indicated with -1 and NaN) with the arithmetic 
mean of the attribute.
c. Create a table with the 5 number summary of all the numeric attributes after treating missing values. ?
'''

#loading the dataframe and viewing it's properties
df = pd.read_csv("cereal.csv")
display(df.describe())
display(df.isnull().sum())
display(df)

#replacing all NaN with -1 and then replacing -1 with column average
df.fillna(-1, inplace=True)
display(df.isnull().sum())
for i in df.columns:
    if i not in ["name", "mfr" ,"type"]:
        df[i].replace(to_replace = -1 , value = df[i].mean() ,inplace = True)
print("\n after replacing -1 with avg")
display(df.describe())
```

### 12. Combining Plots:
○ Create a figure with multiple subplots (line, bar, and scatter plots)

```python
# Set figure size and style
plt.figure(figsize=(15, 15))
sns.set_style("darkgrid")

#generating line graph of variation of calories in various serials
plt.subplot(3, 1, 1)
plt.plot(df["name"], df["calories"], marker="o", linestyle="-", color="g", label="Calories")
plt.xticks(rotation=90)
plt.title("Calories various serieals")
plt.xlabel("cereal Name")
plt.ylabel("calories")

#generating bar plot of avg sugar per manufacturer
plt.subplot(3, 1, 2)
sugar_by_mfr = df.groupby("mfr")["sugars"].mean().sort_values()
sns.barplot(x=sugar_by_mfr.index, y=sugar_by_mfr.values, palette="viridis")
plt.title("Average Sugar Content by Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Average Sugar (grams)")

#generating scatter plot of carbohydrates vs fibre in sugar content
plt.subplot(3, 1, 3)
sns.scatterplot(x=df["carbo"], y=df["fiber"], hue=df["sugars"], size=df["calories"], palette="coolwarm", sizes=(20, 200))
plt.title("Carbohydrates vs. Fiber")
plt.xlabel("carbo")
plt.ylabel("fibre")
plt.tight_layout()
plt.show()
```
