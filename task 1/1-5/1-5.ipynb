```python
from numpy import random
import numpy
array = random.randint(1,50 , size=(3,4))
print("mean is " , numpy.mean(array))
print("median is " , numpy.median(array))
print("standard deviation is " , numpy.std(array))
print("\n")
print("sum of all elements is" , numpy.sum(array))
print("sum of each row is:")
for i in array:
    print(numpy.sum(i))
print("\nreshaped array is:")
numpy.reshape(array , (2,6))
```

```python
array1 = random.randint(1,10,(3,3))
array2 = random.randint(1,10,(3,3))
array_sum = array1+array2
array_sub = array1-array2
array_dot = numpy.dot(array1,array2)
array_cross = numpy.multiply(array1,array2)
array1_transpose = numpy.transpose(array1)
array2_transpose = numpy.transpose(array2)
array1_det = numpy.linalg.det(array1)
array2_det = numpy.linalg.det(array2)

print("array1 is:\n" , array1 , "\narray2 is:" , array2)
print("\n sum is: " , array_sum , "product is: " , array_sub , "dot product is: " , array_dot , "cross product is: ",
      array_cross ,sep = "\n")
print("\n transpose of array1 is:" , array1_transpose , "\n transpose of array2 is:" , array2_transpose , sep = "\n")
print("\n det of array1 is: " , array1_det ,"\n det of array2 is: " , array2_det , sep = "\n" )
print()

if array1_det == 0:
    print("No as determinant is 0 , inverse is not possible for array 1")
else:
    print("inverse of array 1 is")
    array_inverse = numpy.linalg.inv(array1)
    print(array_inverse)
if array2_det == 0:
    print("No as determinant is 0 , inverse is not possible for array 2")
else:
    print("inverse of array 2 is")
    array_inverse = numpy.linalg.inv(array2)
    print(array_inverse)
```

```python
import random
import pandas as pd
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data_dict = {}
for i in days_of_week:
    data_dict[i] = random.randint(-5 , 35)
series = pd.Series(data_dict)

print("the data is:\n" , series)
print("\nmean temperature is:" , series.mean())
print("the maximum temperature is:" , series.max() , "and the day is:" , series.idxmax())
print("the minimum temperature is:" , series.min() , "and the day is:" , series.idxmin())

print()
specific_value = float(input("Enter a temperature:"))
print("temperatures greater than" , specific_value , "are:" , (series[series>specific_value]).values)

print()
series = series * 9 / 5 + 32
print("series in farenheit is:\n" , series)

print()
print("days on which temp is greater than average is:")
mean_temp = series.mean()
d = series > mean_temp
d_index = d.index
for i in d_index:
    if d[i] == True:
        print(i)
```

```python
#opening and loading the dataframe
import warnings
warnings.filterwarnings('ignore')
with open("cereal.csv" , mode= "r" , newline = "") as f:
    df = df = pandas.read_csv(f)
print("The original data given is")
display(df)
df.describe
#creating the 5 number summary : min 25 meidan 75 and max
five_nos_summary_df = df.describe
print("The five number summary is")
display(five_nos_summary_df)

#replacing all -1 with column average
print("\nreplacing -1 with column average")
for i in df.columns:
    if i not in "name mfr type":
        df[i].replace(to_replace = -1 , value = df[i].mean() ,inplace = True)
display(df)

print("\nnew 5 number summary is")
five_nos_summary_df = df.describe
display(five_nos_summary_df)

#removing noisy values and replacing them with the column median
for i in df.columns:
    if i not in "name mfr type":
        q1 = df[i].quantile(0.25)
        q3= df[i].quantile(0.75)
        IQR = q3-q1
        lower_limit = q1*1.5 * IQR
        upper_limit =  q3*1.5 * IQR  
        median = df[i].median()
        for j in df[i]:
            if j<lower_limit or j>upper_limit:
                df[i].replace(to_replace = j, value=median , inplace = True)
print("\nnew 5 number summary after replacing noisy data is:")
five_nos_summary_df = df.describe
display(five_nos_summary_df)
df.shape
```
